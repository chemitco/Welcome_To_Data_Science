```python
df2=pd.DataFrame({'ID':[1,2,3,4],
                'Name':['A','B','C','D']})
df2
```

```python
pd.merge(df1,df2, on='ID') #ID is coomon so we have used on to make common)
```

```python
pd.merge(df2,df1, on='ID') #We have used left and right parameter
```

```python
#Now ID are different
```

```python
pd.merge(df1,df2, on='ID', how='inner')
```

```python
pd.merge(df1,df2, on='ID', how='right')
```

```python
pd.merge(df1,df2, on='ID', how='outer')
```

```python
pd.merge(df1,df2, on='ID', how='outer', indicator=True) #It will give you the information about each row.
```
